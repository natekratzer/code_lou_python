# Code Louisville intro to Pandas

Pandas is a library that allows us to deal with data in a dataframe format. This is really useful for being able to quickly do data analysis. When importing, it is conventional to import as 'pd' which allows it to later be referenced by typing 'pd' rather than 'pandas'

```python
import pandas as pd
import os
```

I'm using the os to find the current working directory. 

```python
os.getcwd()
```

I downloaded the data from here: http://greaterlouisvilleproject.org/deep-drivers-of-change/education/ and saved it in my current working directory. Now I'm using Pandas' read_excel command to read it in and save it as a dataframe named edu_df. 

```python
edu_df = pd.read_excel('GLP-Codebook.xlsx', 'Edu County', index_col=None, na_values=['NA'])
edu_df
```

Pandas makes it easy to get summary statistics for the whole dataframe. Note though, that Pandas guesses what kind of data it is dealing with. A mean year of 2010.5 doesn't make much sense.

```python
edu_df.describe()
```

```python
edu_df['year'] = edu_df['year'].astype('category')
edu_df.describe()
```

```python
edu_df['year'].describe()
```

```python
edu_df['child_per'].describe()
```

```python
edu_df.corr()
```

```python
Pandas .filter method can be used to select a subset of the columns
```

```python
edu_sel_df = edu_df.filter(items = ['city', 'year', 'current', 'child_per', 'bach_plus_per_all', 'per_high_wage'])
edu_sel_df
```

Brack indices can be used to subset by value. In this example we're taking only those cities that are currently peer cities. 

```python
edu_sel_15_df = edu_sel_df[(edu_sel_df.current == 1) & (edu_sel_df.year == 2015)]
edu_sel_15_df
```

```python
edu_sel_15_df
```

A new column can be created simply by naming it after the dataframe inside of a pair of []. A common data manipulation is to add/subtract/divide etc. one column by another one. 

```python
edu_df['bach_plus_race_gap'] = edu_df['bach_plus_per_white'] - edu_df['bach_plus_per_black']
edu_df
```
